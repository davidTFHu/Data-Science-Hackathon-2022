```python
import pandas as pd
import semopy
```

```python
data = pd.read_csv('data.csv')
print(data.head())
```

```python
# Specify the model relations
model_spec = """
  # measurement model
    optUserAction =~ searchInv + testDrive + findDealer
    passive =~ incView + avgPageVis + paidSearchVis + bannerVis + vehHomeVis + galleryVis
    aware =~ bpFinish + bpStart + quickQuote + paymentEst + naturalSearchVis + bookmarkedVis
    explore =~ vehQuote + tradeIn
    consider =~ clickCall + avgSiteTime
  # regressions
    optUserAction ~ passive + aware + explore + consider
"""

# Instantiate the model
model = semopy.Model(model_spec)

# Fit the model using the data
model.fit(data)
```

```python
# Show the results using the inspect method
model.inspect()
```

```python
print(semopy.calc_stats(model))
```
